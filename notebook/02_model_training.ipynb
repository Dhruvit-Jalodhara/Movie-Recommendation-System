```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('/Users/dhruvitjalodhara/programming/ML Practice/Movie Recommendation System/movie dataset/processed/filtered_movies.csv')
ratings = pd.read_csv('/Users/dhruvitjalodhara/programming/ML Practice/Movie Recommendation System/movie dataset/processed/filtered_ratings.csv')
```

```python
ratings = ratings.pivot(index='movieId' , columns='userId' , values='rating')
```

```python
ratings = ratings.fillna(0)
```

```python
ratings
```

```python
# Compress into a CSR Matrix : A Compressed Sparse Row (CSR) matrix completely ignores the zeros in memory and only tracks the exact row/column coordinates of the actual numeric ratings.
from scipy.sparse import csr_matrix

ratings_csr = csr_matrix(ratings.values)
```

```python
print("--- MATRIX ENGINEERING SUCCESSFUL ---")
print(f"Pivot DataFrame Shape : {ratings.shape}")
print(f"Compressed Sparse CSR Shape: {ratings_csr.shape}")
```

```python
## Training model by KNN
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric='cosine' , algorithm='brute' , n_neighbors=20)
model_knn.fit(ratings_csr)
```

```python
# 1. Create a helper series matching movie titles to their matrix row numbers
# We reset the index of ratings to get numerical positions (0, 1, 2...)
matrix_movie_ids = pd.DataFrame({'movieId': ratings.index})

# 2. Merge with the movies dataframe to get the matching titles for those rows
# This ensures perfect row alignment and records the matrix index position
mapping_df = pd.merge(matrix_movie_ids, movies, on='movieId').reset_index()

# 3. Create the final Series: index is 'title', values are the matrix row positions ('index')
movie_to_idx = pd.Series(mapping_df['index'].values, index=mapping_df['title'])

print("--- MAPPING REFERENCE CREATED ---")
print(f"Total mapped movies: {len(movie_to_idx)}")

movie_to_idx
```

```python
search_title = "Death Note: Desu nôto (2006–2007)"

# Step 1: Use your mapping reference to look up that movie's exact row index
try:
	movie_matrix_idx = movie_to_idx[search_title]
	print(f"✅ Found '{search_title}' at Matrix Row Index: {movie_matrix_idx}\n")
except KeyError:
	print(f"❌ '{search_title}' not found. Please double-check the exact spelling/year.")
	movie_matrix_idx = None

if movie_matrix_idx is not None:

	# Step 2: Extract that specific row's data vector from your pivoted ratings dataframe
	movie_vector = ratings.iloc[movie_matrix_idx].values

	# Step 3: Pass vector into kneighbors(), reshaping it into a 2D array (.reshape(1, -1))
	# We request 6 neighbors to get the top 5 recommendations (since index 0 is the movie itself)
	distances, indices = model_knn.kneighbors(
	movie_vector.reshape(1, -1),
	n_neighbors=6
	)

	# Step 4: Flatten the returned arrays to make them easy to loop through
	flat_distances = distances.flatten()
	flat_indices = indices.flatten()

	idx_to_movie = {v: k for k, v in movie_to_idx.items()}

	# Step 5: Loop through neighbor indices, look up actual titles, and print results
	print(f"Recommendations for '{search_title}':")
	print("-" * 50)

	# Start loop from 1 to skip the searched movie itself
	for i in range(1, len(flat_indices)):
		neighbor_row_idx = flat_indices[i]
		neighbor_distance = flat_distances[i]

		# Reverse lookup the title from your movie_to_idx series
		#recommended_title = movie_to_idx[movie_to_idx == neighbor_row_idx].index[0]

		recommended_title = idx_to_movie[neighbor_row_idx]

		# Convert distance to a similarity score (1 - distance)
		similarity_score = 1 - neighbor_distance
		print(f"{i}. {recommended_title} (Similarity: {similarity_score:.2%})") 
```

```python
## Truncated SVD
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=50 , random_state=42)
movie_features_svd = svd.fit_transform(ratings_csr)
movie_features_svd
```

```python
from sklearn.metrics.pairwise import cosine_similarity

svd_sim_matrix = cosine_similarity(movie_features_svd)
svd_sim_matrix
```

```python
# Step 0: Look up the row index for your target movie using the mapping reference
target_movie = "Sultan (2016)"
movie_idx = movie_to_idx[target_movie]

# Step 1: Extract that movie's entire similarity score row from svd_sim_matrix
similarity_row = svd_sim_matrix[movie_idx]

# Step 2: Sort the scores from highest to lowest using np.argsort and reversing it
sorted_indices = np.argsort(similarity_row)[::-1]

# Step 3: Grab the top 6 indices (slicing from index 1 to 6 to skip the target movie itself)
top_6_neighbors = sorted_indices[1:6]

# Step 4: Loop through neighbor indices, reverse-lookup titles, and print them out
print(f"SVD Recommendations for '{target_movie}':")
print("-" * 60)

# Create a fast reverse lookup dictionary from your existing mapping Series
idx_to_movie_dict = {v: k for k, v in movie_to_idx.items()}

for rank, neighbor_idx in enumerate(top_6_neighbors, 1):

	# Reverse lookup the title from your movie_to_idx series
	#recommended_title = movie_to_idx[movie_to_idx == neighbor_idx].index[0]

	recommended_title = idx_to_movie_dict[neighbor_idx]
	
	# Extract the similarity score
	similarity_score = similarity_row[neighbor_idx]
	print(f"{rank}. {recommended_title} (SVD Match: {similarity_score:.2%})") 
```

```python
## Non-Negative Matrix Factorization (NMF)
from sklearn.decomposition import NMF

nmf = NMF(n_components=50, init='random', random_state=42)
movie_features_nmf = nmf.fit_transform(ratings_csr)
movie_features_nmf
```

```python
from sklearn.metrics.pairwise import cosine_similarity

nmf_sim_matrix = cosine_similarity(movie_features_nmf)
nmf_sim_matrix
```

```python
# Step 0: Look up the row index for your target movie using the mapping reference
target_movie = "Avengers: Infinity War - Part II (2019)"

try:
    movie_idx = movie_to_idx[target_movie]
    print(f"✅ Found '{target_movie}' at Matrix Row Index: {movie_idx}\n")
    
    # Step 1: Extract that movie's entire similarity score row from nmf_sim_matrix
    similarity_row = nmf_sim_matrix[movie_idx]

    # Step 2: Sort the scores from highest to lowest using np.argsort and reversing it
    sorted_indices = np.argsort(similarity_row)[::-1]

    # Step 3: Grab the top 6 indices (slicing from index 1 to 6 to skip the target movie itself)
    top_6_neighbors = sorted_indices[1:6]

    # Step 4: Loop through neighbor indices, reverse-lookup titles, and print them out
    print(f"NMF Recommendations for '{target_movie}':")
    print("-" * 60)

    # Create your optimized reverse lookup dictionary
    idx_to_movie_dict = {v: k for k, v in movie_to_idx.items()}

    for rank, neighbor_idx in enumerate(top_6_neighbors, 1):
        # Look up the title instantly via your dictionary map
        recommended_title = idx_to_movie_dict[neighbor_idx]
        
        # Extract the similarity score
        similarity_score = similarity_row[neighbor_idx]
        
        print(f"{rank}. {recommended_title} (NMF Match: {similarity_score:.2%})")

except KeyError:
    print(f"❌ '{target_movie}' not found in your filtered dataset.")
    print("This usually means the movie didn't meet your minimum rating cut-off threshold.") 
```

```python
## K-Means Clusturing
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=30, random_state=42, n_init='auto')
cluster_labels = kmeans.fit_predict(movie_features_svd)
cluster_labels
```

```python
# Assign directly to mapping_df for perfect index alignment
mapping_df['cluster'] = cluster_labels

mapping_df
```

```python
# Step 0: Direct lookup using your exact mapping dictionary/Series keys
search_title = "Kung Fu Panda (2008)"

if search_title in movie_to_idx.index:
	# Get the row index first
	matrix_row_idx = movie_to_idx[search_title]
	# Extract the cluster number directly from that row in mapping_df
	assigned_cluster = mapping_df.loc[matrix_row_idx, 'cluster']
	print(f"🎬 Movie: {search_title} ➡️ Cluster: {assigned_cluster}\n")
	# Step 1: Filter the dataframe to isolate this cluster group
	cluster_group = mapping_df[mapping_df['cluster'] == assigned_cluster]
	# Step 2: Grab and print 5 random movie titles from the same cluster
	print(f"--- 10 Random Neighbors from Cluster {assigned_cluster} ---")
	print(cluster_group['title'].sample(5).to_string(index=False))

else:
	print(f"❌ '{search_title}' not found. Let's look for a partial match:")
	# Fallback to show partial matches if the year or text is slightly off
	matches = mapping_df[mapping_df['title'].str.contains("Avengers", case=False, na=False)]['title'].head()
	print(matches.to_string(index=False)) 
```

```python
def calculate_batch_mse(W, H, original_csr, batch_size=1000):
	"""
	Calculates Mean Squared Error in small, memory-safe chunks.
	W: Movie features matrix (e.g., movie_features_svd)
	H: Components matrix (e.g., svd.components_)
	original_csr: Your compressed sparse matrix (ratings_csr)
	"""

	total_squared_error = 0.0
	num_movies = original_csr.shape[0]
	total_elements = original_csr.shape[0] * original_csr.shape[1]

	# Process the matrix in small chunks of rows
	for start_idx in range(0, num_movies, batch_size):
		end_idx = min(start_idx + batch_size, num_movies)

		# 1. Get the dense chunk of original ratings directly from the sparse matrix
		actual_chunk = original_csr[start_idx:end_idx].toarray()

		# 2. Reconstruct ONLY this small chunk of rows
		predicted_chunk = np.dot(W[start_idx:end_idx], H)

		# 3. Sum up the squared differences for this chunk
		total_squared_error += np.sum((actual_chunk - predicted_chunk) ** 2)

	# Final step: Divide the grand total by the total number of cells
	return total_squared_error / total_elements

# --- RUN EXECUTION MEMORY-SAFELY ---
print("Calculating TruncatedSVD Error...")
svd_mse = calculate_batch_mse(movie_features_svd, svd.components_, ratings_csr)

print("Calculating NMF Error...")
nmf_mse = calculate_batch_mse(movie_features_nmf, nmf.components_, ratings_csr)

print("\n--- RECONSTRUCTION ERROR CALCULATED ---")
print(f"TruncatedSVD Reconstruction MSE: {svd_mse:.6f}")
print(f"NMF Reconstruction MSE: {nmf_mse:.6f}")
```

```python
# Select a random sample of 200 movie indices to test our models across
np.random.seed(42)
total_unique_movies = ratings.shape[0]
sample_indices = np.random.choice(total_unique_movies, size=200, replace=False)

# Sets to store unique recommended movie row indices
knn_recommended_set = set()
svd_recommended_set = set()
nmf_recommended_set = set()

print("Calculating coverage for 200 sample movies...")

for idx in sample_indices:
	# --- KNN Coverage ---
	# Using ratings_csr[idx] directly avoids unpack crashes and preserves memory
	_, indices_knn = model_knn.kneighbors(ratings_csr[idx], n_neighbors=11)
	knn_recommended_set.update(indices_knn.flatten()[1:]) # skip self at index 0
	# --- SVD Coverage ---
	svd_scores = svd_sim_matrix[idx]
	top_svd = np.argsort(svd_scores)[::-1][1:11] # grab positions 1 through 10
	svd_recommended_set.update(top_svd)
	# --- NMF Coverage ---
	nmf_scores = nmf_sim_matrix[idx]
	top_nmf = np.argsort(nmf_scores)[::-1][1:11] # grab positions 1 through 10
	nmf_recommended_set.update(top_nmf)

# Convert sets to catalog coverage percentages
knn_coverage = (len(knn_recommended_set) / total_unique_movies) * 100
svd_coverage = (len(svd_recommended_set) / total_unique_movies) * 100
nmf_coverage = (len(nmf_recommended_set) / total_unique_movies) * 100

print("\n--- ITEM CATALOG COVERAGE CALCULATED ---")
print(f"KNN Catalog Coverage: {knn_coverage:.2f}%")
print(f"TruncatedSVD Catalog Coverage: {svd_coverage:.2f}%")
print(f"NMF Catalog Coverage: {nmf_coverage:.2f}%")
```
